### IMDB Top rated 100 movies

```python

import requests 
from bs4 import BeautifulSoup
```

```python
url = requests.get("https://www.imdb.com/search/title/?count=100&groups=top_100&sort=user_rating")
```

```python
url
```

```python
imdb=BeautifulSoup(url.content)
imdb
```

```python
Name = imdb.find_all('h3',class_="lister-item-header")
Name
```

```python

Movie_Name = []

for i in Name:
    Movie_Name.append(i.text.replace("\n",""))
    
Movie_Name = [i.split('.')[1] for i in Movie_Name]
Movie_Name = [i.split('(')[0] for i in Movie_Name]
Movie_Name
```

```python
R_year = imdb.find_all("span",class_="lister-item-year text-muted unbold")
R_year
```

```python
year=[]
for i in R_year :
    year.append(i.text.replace("\n",""))
    
year = [i.split('(')[-1] for i in year] 
year = [i.split(')')[0] for i in year]
```

```python
year
```

```python
rating = imdb.find_all("div",class_="inline-block ratings-imdb-rating")
rating
```

```python
ratings = []

for i in rating:
    ratings.append(i.text.replace("\n",""))
    
ratings
```

```python
movies = {"Name":Movie_Name,"Year":year,"Rating":ratings}
```

```python
import pandas as pd


Top_100_movies = pd.DataFrame(movies)
```
