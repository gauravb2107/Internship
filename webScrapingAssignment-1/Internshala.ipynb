# Internshala

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = requests.get("https://internshala.com/fresher-jobs")
url
```

```python
soup = BeautifulSoup(url.content)
soup
```

```python
title = soup.find_all("div",class_="heading_4_5 profile")
title
```

```python
Titles = []

for i in title:
    Titles.append(i.text.replace("\n",""))
```

```python
Titles
```

```python
name = soup.find_all("a",class_="link_display_like_text")
name
```

```python
Company_Name = []

for i in name:
    Company_Name.append(i.text.strip())
```

```python
Company_Name
```

```python
loc = soup.find_all("p",id="location_names")
loc
```

```python
location = []

for i in loc:
    if i.text is None:
        location.append("No data")
    else:
        location.append(i.text.replace("\n",""))
        
location
```

```python
apply_date=[]
job_info = []
CTC=[]
info = soup.find_all("div",class_="item_body")
for i in info:
    job_info.append(i.text.replace("\n","").replace("\xa0",""))

for i in range(2,len(job_info),3):
    apply_date.append(job_info[i].strip())
for i in range(1,len(job_info),3):
    CTC.append(job_info[i].strip())
```

```python
import pandas as pd

jobs = pd.DataFrame({"Title":Titles,"Company":Company_Name,"Location":location,"CTC":CTC,"Apply_Date":apply_date})

jobs
```
