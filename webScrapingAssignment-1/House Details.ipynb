```python
soup = BeautifulSoup(url.content)
soup
```

```python
houses = soup.find_all("h2", class_="heading-6 font-semi-bold nb__1AShY")

house=[]
for i in houses:
    house.append(i.text)
```

```python
house
```

```python
loc = soup.find_all('div',class_="nb__2CMjv")

location = []
for i in loc:
    location.append(i.text)
    
location
```

```python
area = soup.find_all('div',class_="nb__3oNyC")

Area=[]
for i in area:
    Area.append(i.text)
    
Area
```

```python
full_info = []
EMI=[]
price=[]

detail = soup.find_all('div',class_="font-semi-bold heading-6")
for i in detail:
    full_info.append(i.text)

#EMI
for i in range(1,len(full_info),3):
    EMI.append(full_info[i])

# price
for i in range(2,len(full_info),3):
    price.append(full_info[i])
```

```python
import pandas as pd

house_details = pd.DataFrame({"House":house,"Location":location,"Area":Area,"Price":price,"EMI":EMI})

house_details
```
