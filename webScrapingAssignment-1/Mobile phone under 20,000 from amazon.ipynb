# Mobile phone under 20,000 from amazon

```python
import requests
from bs4 import BeautifulSoup
```

```python
page=requests.get("https://www.amazon.in/s?k=mobile+phones+under+20000&ref=nb_sb_noss")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
product=soup.find_all('span',class_="a-size-medium a-color-base a-text-normal")
product
```

```python
Product=[]  
for i in product:
    Product.append(i.get_text())
Product
```

```python
price=soup.find_all('span',class_="a-price-whole")
price
```

```python
Price=[]  
for i in price:
    Price.append(i.get_text())
Price
```

```python
url=soup.find_all('img',class_="s-image")
url
```

```python
Image=[]
for i in url:
    Image.append(i['src'])
Image
```

```python
rating=soup.find_all('span',class_="a-icon-alt")
rating
```

```python
Rating=[]  
for i in rating:
    Rating.append(i.get_text().replace('amp;',' '))
Rating
```

```python
print(len(Product),len(Price),len(Image),len(Rating))
```

```python
import pandas as pd
mobile=pd.DataFrame({})
mobile['Product_Name']=Product
mobile['Price']=Price
mobile['Image_Url']=Image
mobile['Average_Rating']=Rating[:16]
```
