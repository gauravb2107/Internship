### Wikipedia all header Tags

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = requests.get('https://en.wikipedia.org/wiki/Main_Page')
```

```python
url
```

```python
content=BeautifulSoup(url.content)
```

```python
content
```

```python
tags = content.find_all(['h1','h2','h3','h4','h5','h6'])
```

```python
tags
```

```python
header = []

for i in tags:
    header.append(i.text.replace("\n",""))
    
header
```
